```python
binary_cols = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 0, 'F': 0, 'M': 1, 'U': 1, 'R': 0, 'LE3': 1, 'GT3': 0, 'T': 1, 'A': 0, 'GP': 1, 'MS': 0})
```

```python
df['Mjob'] = df['Mjob'].replace({'teacher': 0, 'health': 0, 'services': 0, 'at_home': 1, 'other': 1})
```

```python
df['Fjob'] = df['Fjob'].replace({'teacher': 0, 'health': 0, 'services': 0, 'at_home': 1, 'other': 1})
```

```python
reason_dummies = pd.get_dummies(df['reason'], prefix='reason')
df = pd.concat([df, reason_dummies], axis=1)
df = df.drop('reason', axis=1)
```

```python
df['guardian'] = df['guardian'].replace({'mother': 0, 'father': 0, 'other': 1})
```

```python
df= df.drop(['G1', 'G2'], axis=1)
```

```python
df
```

```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
X = df.drop('G3', axis=1)
y = df['G3']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
coefficients = pd.DataFrame({'Variable': X.columns, 'Coefficient': model.coef_})
print(coefficients)
print('Intercept:', model.intercept_)
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
svm_model = SVR(kernel='linear')
svm_model.fit(X_train_scaled, y_train)
y_pred = svm_model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
coefficients = svm_model.coef_
print('Coefficients:', coefficients)
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
knn_regressor = KNeighborsRegressor(n_neighbors=5) #3,5,7
knn_regressor.fit(X_train , y_train)
y_pred = knn_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
regressor = DecisionTreeRegressor(max_depth=7) #3,5,7
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
rf_model = RandomForestRegressor(n_estimators=100,max_depth=7, random_state=42) #
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.ensemble import AdaBoostRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
ada_boost = AdaBoostRegressor(n_estimators=500,learning_rate=0.1, random_state=42)
ada_boost.fit(X_train, y_train)
y_pred = ada_boost.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
gb_regressor = GradientBoostingRegressor(n_estimators=500, learning_rate=0.1,max_depth=3,random_state=42)
gb_regressor.fit(X_train, y_train)
predictions = gb_regressor.predict(X_test)
mse = mean_squared_error(y_test,predictions)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test,predictions)
print(f'R-squared: {r2}')
```

```python
from xgboost import XGBRegressor
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
xgb_model = XGBRegressor(n_estimators=500, learning_rate=0.1,max_depth=3,random_state=42)
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test,y_pred)
print(f'R-squared: {r2}')
```
